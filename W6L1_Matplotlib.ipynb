# How to use matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['A', 'B', 'C', 'D', 'E']
x = np.array([50, 20, 15, 10, 5])
plt.pie(x, labels=labels)
plt.show()
```

```python
# 나눔글꼴 설치 코드, 실행 후 런타임 메뉴, "세션 다시 시작"
!sudo apt-get install -y fonts-nanum
!sudo fc-cache -fv
!rm ~/.cache/matplotlib -rf
```

```python
#글꼴 설치하고 런타임(세션 다시 시작) 후 matplotlib 다시 import 해야 한글 보임

plt.rc('font', family='NanumBarunGothic')
plt.pie(x, labels=labels, shadow=True, startangle=90, autopct='%0.1f %%')
plt.title('시장점유율')
plt.show()
```

```python
colors = ['#8dd3c7', '#ffffb3', '#bebaba', '#fb8072', '#80b1d3']
#plt.rc('font', family='NanumBarunGothic')
plt.pie(x, labels=labels, colors=colors, startangle=90, autopct='%0.1f %%')
plt.title('시장점유율')
plt.show()
```

```python
colors = ['#8dd3c7', '#ffffb3', '#bebaba', '#fb8072', '#80b1d3']
plt.rc('font', family='NanumBarunGothic')
plt.pie(x, labels=labels, colors=colors, startangle=90, autopct='%0.1f %%', counterclock=False)
#0도에서 90도 왼쪽으로 간 다음, counterclock=False 이므로 시계방향 (기본은 반시계 방향인 True값)
plt.title('시장점유율')
plt.show()
```

```python
plt.pie(x, labels=labels, colors=colors, startangle=90, autopct='%0.1f %%', explode=[0, 0.1, 0.3, 0.3, 0])
#explode : 특정 데이터에 집중해달라, 피자조각 빼내듯이
#0은 안빼내는거, 숫자 커질수록 더 많이 떨어짐 (예: 0.1,0.3,1.0 ... 해보기)
#autopct 빼니까 숫자가 안나옴
plt.title('시장점유율')
plt.show()
```

```python
plt.pie(x, labels=labels, colors=colors, startangle=90,
        autopct='%0.1f %%', explode=[0, 0, 0.3, 0, 0],
        wedgeprops={'edgecolor':'black', 'linestyle':'--' }, pctdistance=1.2, labeldistance=0.7)

#wedge : pie chart에서 하나의 데이터를 나타내고 단일 색깔로 되어있는 영역
#wedgeprops : 어떤 속성을 더 줄까, wedge property의 약자
#wedge의 파라미터를 딕셔너리로 전달
#검정색 선인데 점선.linestyle':'--'
#실선 : linestyle':'-'

#pctdistance(percent distance네) : %(값)에 관한 위치 조정, 기본은 0.5(위의 예제) / 1.0 이면 반지름을 1로 간주,
#반지름 대비 어느정도 위치에 배치할것 인가를 비율로 결정
#1.0 보다 큰 1.2 같은 값이면 파이차트를 나가서 적힘

#labeldistance : 이름(label)에 관한 위치 조정 / 1보다 작은 값을 넣어서 파이차트 안에 적음
plt.title('시장점유율')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
```

```python
d1 = np.random.randint(low=1, high=11, size=50) #1~10까지 50개
d2 = d1 + np.random.randint(1, 5, size=d1.size) #2~14까지
data = np.column_stack((d1, d2)) #두 개의 배열 d1과 d2를 열(column) 기준으로 나란히 붙여서 2차원 배열(행렬)로 만든다
data
```

밑의 코드 추가 공부

hist(data, ...) 형태로 쓰면:
data의 각 열(즉, d1과 d2)을 따로따로 히스토그램으로 그림

즉, x축에는 값들, y축에는 그 값들의 빈도수(횟수)


bins=np.arange(data.min(), data.max())에서

bins: 히스토그램의 구간을 지정

np.arange(data.min(), data.max()):

전체 데이터에서 가장 작은 값부터 가장 큰 값 직전까지의 정수 간격 구간들을 만든다는 뜻

예: data.min() = 1, data.max() = 14이면 bins = [1, 2, ..., 13]

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
ax1.scatter(x=d1,y=d2, marker='o', c='r', edgecolor='b')
ax1.set_title('Scatter: $x$ versus $y$')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')

ax2.hist(data, bins=np.arange(data.min(), data.max()),
label=('x', 'y'))
ax2.legend(loc=(0.65, 0.8))
ax2.set_title('Frequencies of $x$ and $y$')
ax2.yaxis.tick_right()
```

#Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(3) # 0, 1, 2
years = ['2017', '2018', '2019']
values = [100, 400, 900]
```

```python
myDF = pd.DataFrame({'Year':years, 'Score':values})
myDF.head() #처음 5개의 행만 보여주는 메서드
```

```python
#데이터 프레임에서 기능 제공
myDF.plot.bar(x='Year', y='Score', rot=0) # rotation
#rot = 0 : x의 눈금의 값이 x축과 수평을 이룸
#rot 으로 45, 90 해보기
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots() #명시적 호춯

ax.bar(myDF.Year, myDF.Score)
ax.set_title('TestScores')
ax.set_xlabel('Year')
ax.set_ylabel('Score')

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

#데이터 정의
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange'] #데이터 개수만큼 지정
#_red : 앞에 범례에 나온 데이터를 다시 사용하려고 할 때 _를 사용해야지 범례에 추가 생성되지 않는다.
bar_colors = ['red', 'tab:blue', 'tab:red', 'tab:orange'] # tableau
#tableau 에서 색깔 정의한거 사용하려면 : tab:원하는색 / 그냥 red 해도 ㄱㅊ
ax.bar(fruits, counts, label=bar_labels, color=bar_colors)
#label='some label'은 막대 전체에 대해 하나의 범례 항목을 지정

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
actors = pd.read_csv('actors.csv')
actors
```

```python
# plotting a line graph
#1980년 대비 2015년이 계속 증가
print("Line graph: ")
plt.plot(movies_by_year["Year"], movies_by_year["Number of Movies"])
plt.show()


# plotting a scatter plot
print("Scatter Plot: ")
plt.scatter(actors["Number of Movies"], actors["Total Gross"])
#Total Gross : 각 배우의 출연작들의 총 수익(박스오피스 총액)
plt.show()
```

Let’s draw a complex(?) graph

```python
# plot style
#linewidth : 라인두께
plt.plot(movies_by_year["Year"], movies_by_year["Number of Movies"], 	linewidth=3.0, label='# of Movies')
plt.plot(movies_by_year["Year"], movies_by_year["Total Gross"], 	linewidth=2.0, label='Total Gross')
plt.title("Example: Meaningless Comparsion") #비교 의미없다
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.legend() #범례 표현
plt.axis([1979, 2016, 0, 12000]) #x축 범위 : 1979~2016, y축 범위 : 0~12000
plt.grid(axis = 'y', color = 'purple', linestyle = '--', linewidth = 1)
# 보기 좋게 눈금을 자동 정렬 (1979,2016도 포함됨)
plt.show()
```

```python
# oo style , explicit 하게 생성
fig, ax = plt.subplots(figsize=(5, 3), layout='constrained')
#layout='constrained'란? 레이아웃 자동 조절 기능 , 그래프 내의 요소들이 겹치지 않도록 자동으로 **여백(padding)**을 계산해서 배치
ax.plot(movies_by_year["Year"], movies_by_year["Number of Movies"], 	linewidth=3.0, label='# of Movies')
ax.plot(movies_by_year["Year"], movies_by_year["Total Gross"], 	linewidth=2.0, label='Total Gross')
ax.set_title("Example: Meaningless Comparsion")
ax.set_xlabel("Year")
ax.set_ylabel("Number of Movies")
ax.legend()
ax.set_xlim([1979,2016]) #다른부분
ax.set_ylim([0,12000])
#ax.grid(axis = 'y', color = 'purple', linestyle = '--', linewidth = 1)
ax.grid(True, linestyle = '--') #grid의 visible이 원래는 None -> True
plt.show()
```
